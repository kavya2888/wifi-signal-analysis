```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for better visuals
sns.set(style="whitegrid")
```

```python
# Load the Wi-Fi signal data
data = pd.read_csv("wifi_data.csv")

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())

# Basic statistics
print("\nBasic statistics of signal strength:")
print(data['signal_strength'].describe())
```

```python
# Pivot the data to create a 2D grid for the heatmap
grid = data.pivot(index="y", columns="x",  values="signal_strength")

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(grid, cmap="viridis", annot=True, fmt=".1f", cbar_kws={'label': 'Signal Strength (dBm)'})
plt.title("Wi-Fi Signal Strength Heatmap")
plt.xlabel("X Coordinate (meters)")
plt.ylabel("Y Coordinate (meters)")
plt.show()
```

```python
# Scatter plot of signal strength
plt.figure(figsize=(8, 6))
plt.scatter(data['x'], data['y'], c=data['signal_strength'], cmap="viridis", s=100)
plt.colorbar(label="Signal Strength (dBm)")
plt.title("Wi-Fi Signal Strength at Different Locations")
plt.xlabel("X Coordinate (meters)")
plt.ylabel("Y Coordinate (meters)")
plt.grid(True)
plt.show()
```
